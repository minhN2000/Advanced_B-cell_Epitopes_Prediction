## Embedding proteins in `.fasta` file
We use [ProtTrans](https://github.com/agemagician/ProtTrans) to embed our protein sequences (including any length) in residue level.

```python
EMBEDDING_INPUT_FILE  = './datasets/fasta/non_human_proteins_blind_set.whole.fasta'
EMBEDDING_OUTPUT_NAME = 'blind'
Embedder = embedding(in_file=EMBEDDING_INPUT_FILE,
                     out_file = EMBEDDING_OUTPUT_NAME, 
                     level='residue', 
                     embed='prott5')

# after running this the embedding will be stored at ./datasets/embedding/{EMBEDDING_OUTPUT_NAME}_residue_prott5.csv
```
